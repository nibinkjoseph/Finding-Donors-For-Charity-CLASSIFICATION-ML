Increasing the accuracy using HyperTuning of decision Tree

Gridsearch CV train on every parameter and it gives the best parameter 

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters={'min_samples_split':range(10,500,20),'max_depth':range(1,20,2),'criterion':['gini','entropy']}
clf_tree=DecisionTreeClassifier()
clf=GridSearchCV(clf_tree,parameters,cv=5)
clf.fit(X_train,y_train)
```

```python
clf.best_estimator_
```

```python
clf_predict=clf.predict(X_test)
```

```python
accuracy_score(clf_predict,y_test)
```
